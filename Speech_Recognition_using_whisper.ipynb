```python
import whisper
model = whisper.load_model("base")
```

```python
def translate_text (model, audio_path):
    result = model.transcribe(audio_path)
    return result["text"]

# Define the function to add the audio path
def add_audio_path(audio_file):
    return "/content/audios/" + audio_file + ".flac"
```

```python
from tqdm import tqdm
```

```python
tqdm.pandas(desc="Adding Audio Path")
df['full_audio_path'] = df['audio_file'].progress_apply(add_audio_path)
```

```python
df.head()
```

```python
tqdm.pandas(desc="Translating Text")
df['translated_text'] = df['full_audio_path'].progress_apply(lambda x: translate_text(model, x))
```

```python
df.to_csv('/content/whisper_text_index.csv', index=True)
```
