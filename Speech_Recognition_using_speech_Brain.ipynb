```python
import torchaudio
import speechbrain as sb
from IPython.display import Audio
from speechbrain.dataio.dataio import read_audio
from speechbrain.pretrained import SpeakerRecognition
from speechbrain.pretrained import EncoderDecoderASR
from speechbrain.pretrained import SepformerSeparation as separator
from tqdm import tqdm
```

```python
asr_model = EncoderDecoderASR.from_hparams(source="speechbrain/asr-crdnn-rnnlm-librispeech", savedir="pretrained_models/asr-crdnn-rnnlm-librispeech")
```

```python
df = df.iloc[1500:1600]
```

```python
print(df)
```

```python
def translate_text (asr_model, audio_path):
    result = asr_model.transcribe_file(audio_path)
    return result

# Define the function to add the audio path
def add_audio_path(audio_file):
    return "/content/audios/" + audio_file + ".flac"
```

```python
tqdm.pandas(desc="Adding Audio Path")
df['full_audio_path'] = df['audio_file'].progress_apply(add_audio_path)
```

```python
df.head()
```

```python
tqdm.pandas(desc="Translating Text")
df['translated_text'] = df['full_audio_path'].progress_apply(lambda x: translate_text(asr_model, x))
```
